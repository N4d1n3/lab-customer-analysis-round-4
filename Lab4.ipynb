```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
data = pd.read_csv('marketing_customer_analysis.csv')
```

```python
numericals = data.select_dtypes(include=[np.number])
numericals.head()
```

```python
categoricals = data.select_dtypes(include=[np.object])
categoricals.head()
```

```python
sns.histplot(numericals, bins=12)
```

```python
sns.histplot(numericals['Income'],bins=10)
plt.show()
```

```python
for i in range(numericals.shape[1]):
    sns.histplot(numericals[numericals.columns[i]], bins=100)
    plt.show()
```

```python
sns.heatmap(numericals.corr().round(2),vmin=-1, vmax=1, center=0, annot = True,square=True,
cmap='seismic')
```
